#### counting in a dictionary
###### we can count the number of repitions of a particular value in list by using a dictionary

```python
list1 = [1,2,3,3,2,1,2,1,3,2,1,2,3]
dict1= {"1":0, "2": 0,"3":0}

for items in list1:
    if items ==1:
        dict1["1"] +=1
    elif items ==2:
        dict1["2"] +=1
    else:
        dict1["3"] +=1
dict1
```

```python
def leng(i):
    c =0
    for item in i:
        c+=1
    return c
```

```python
def sum_list(i):
    add =0
    for item in i:
        add+=item
    return add
```

```python
class rectangle:
    def __init__(self , length, breadth):
        self.breadth = breadth
        self.length = length
    def area(self):
        return self.heigth +self.breadth
    
    
    
class square(rectangle):
    def __init__(self , length, breadth):
        super().__init__(length, breadth)
        
    def area(self):
        if self.length == self.breadth:
            return self.length * self.breadth
        else:
            return self.length * self.length
        
        
class cuboid(rectangle):
    def __init__(self , length, breadth, height):
        super().__init__(length, breadth)
        self.height = height
        
    def volume(self):
        return self.length * self.breadth * self.height
    
    
    
square_obj = square(5,6)
square_obj.area()
cuboid_obj = cuboid(2,3,4)
cuboid_obj.volume()

    
            
```
