# Selecting a column from a pandas DataFrame


<h3>Since our axis in pandas have labels, we can select data using those labels.
Unlike in NumPy, we donot need to know the exact index location of a pandas dataframe.
To do this, we can use the DataFrame.loc[] attribute. The syntax for DataFrame.loc[] is:</h3>

df.loc[row_label, column_label] <br>
We can use the following shortcut to select a single column:

df["column_name"]<br>
This style of selecting columns is very common.

```python
import pandas as pd
weather_df = pd.read_csv('weather_data_nan.csv')
weather_df.head()
```

```python
## .loc method 
weather_df.loc[:, 'event']
```

```python
## shortcut method
weather_df['event']
```
