```python
teams_dict = {'Mumbai Indians' : "MI", 'Kolkata Knight Riders' : "KKR", 'Rajasthan Royals' : "RR", 
              'Royal Challengers Bangalore' : "RCB",
            'Kings XI Punjab' : "KXIP",'Chennai Super Kings' : "CSK",'Sunrisers Hyderabad' : "SRH",'Delhi Capitals': "DC" }
```

```python
top8_ipl_teams[['Team1', 'Team2', 'TossWinner' , 'WinningTeam']] = top8_ipl_teams[['Team1', 'Team2', 'TossWinner' , 'WinningTeam']].replace(teams_dict)
```

```python
top8_ipl_teams.head()
```

```python
team_stats_dict = {}
for row in top8_ipl_teams.values:   ## .values le numpy array ma convert garxa
    teams = row[-1].split("Vs")
    for team in teams:
        team = team.strip()
        if team not in team_stats_dict:
            team_stats_dict[team] = 1
        else:
            team_stats_dict[team] += 1
   
```

```python
def univariate_barplot(df, title, xlabel, ylabel, figsize, colors):

    plt.figure(figsize=figsize)
    columns = list(df.columns)
    ax = sns.barplot(x=columns[0],y=columns[1],data = df, palette = colors)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    for value in ax.patches:
        ax.text(value.get_x(), value.get_height() + 0.5, value.get_height())
    plt.show()
```

```python


colors = ['blue', 'yellow', 'gray', 'green', 'pink','red',
                'pink', 'skyblue']
univariate_barplot(team_stats[['Teams', 'Matches Won']].sort_values('Matches Won',
                                                                    ascending = False),
                   'Matches Won By IPL Teams',
                  xlabel = 'Top 8 Ipl Teams', ylabel = "Matches Won", 
                  figsize = (12,8), colors = colors)

```

```python
toss_decider = pd.crosstab(top8_ipl_teams['TossWinner'], top8_ipl_teams['WinningTeam'], margins = True)
toss_decider.head()
```

```python
# Chi-Square
```

```python
import scipy.stats as stats
## alpha = level of significance
alpha = 0.05
H0 = "H0 : There is no relationship between winning toss and winning games"
H1 = "H1 : Winning a t0ss is an advantage in winning games "
test_stats = stats.chi2_contingency(observed = toss_decider)
p_value = test_stats[1]
if p_value <= alpha:
    print(H1)
else:
    print(H0)
```

## home ground is winner deciding or not

```python
## league_game ma matra home rw away hunxa
removed_match = ['Elimination Final','Semi Final', 'Qualifier', 'Qualifier 1', 'Eliminator', 'Qualifier 2', 'Final' ]
```

```python
league_matches = top8_ipl_teams.loc[~top8_ipl_teams['MatchNumber'].isin(removed_match)]
league_matches.head()
```

```python
home_decider = pd.crosstab(league_matches['Team1'] , league_matches['WinningTeam'], margins = True)
home_decider.head()
```

```python
import scipy.stats as stats
## alpha = level of significance
alpha = 0.05
#Null Hypothesis
H0 = "H0 : Playing a game in home is not an advantage for home Team"
#Alternative hypothesis
H1 = "H1 : Playing a game in home is not an advantage for home Team "
test_stats = stats.chi2_contingency(observed = toss_decider)
p_value = test_stats[1]
if p_value <= alpha:
    print(H1)
else:
    print(H0)
```
